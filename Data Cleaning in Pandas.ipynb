# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\capra\OneDrive\Alex The Analyst\Pandas\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates() # Removes duplicates - very easy! Assign it so it can be saved 
df
```

```python
df = df.drop(columns = "Not_Useful_Column") # Remove the column. Assign it so it can be saved ! 
# you can also do - df.drop(columns = "Not_Useful_Column", inplace=True)
df
```

```python
df["Last_Name"].str.strip() # by default, it takes out whitespace 
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...") # lstrip only does it once at a time. 
# df["Last_Name"] will only apply the df changes on the last name 
df
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/") # do it all at once using Regex and just strip
# 123 all numbers, . all dots, _ all underscores, / all slashes
df
```

```python
# all numbers must be 123-456-7890
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True) # in Regex, ^ is caret. caret means match everything except these....a-z or A-Z or 0-9
# anything except those, replace with a blank
df
```

```python
df["Phone_Number"].apply(lambda x: x[0:3]+'-' + x[3:6]+'-' + x[6:9]) #Error:  it doesnt work because the NaN defaults to str, so you want either a 
# string or a number, but in this case, you need only a string because numbers cannot do with this function 
```

```python
df["Phone_Number"]= df["Phone_Number"].apply(lambda x: str(x)) # convert to a string, each entry 
df
```

```python
df["Phone_Number"]= df["Phone_Number"].apply(lambda x: x[0:3]+'-' + x[3:6]+'-' + x[6:9]) # works now 
df
```

```python
df["Phone_Number"]= df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"]= df["Phone_Number"].str.replace('Na--', '')
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(",", n=2, expand = True)
df
```

```python
df["Paying Customer"]= df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"]= df["Paying Customer"].str.replace('No', 'N')
df
```

```python
df["Do_Not_Contact"]= df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"]= df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','') #Or you can do one below this 
df
```

```python
# Or you can do this 
df = df.fillna('') # all fields that have no value will become blank 
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
df
```

```python
df= df.reset_index(drop=True)
df
```
